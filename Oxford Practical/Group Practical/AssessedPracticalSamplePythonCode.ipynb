```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
```

# Load dataset

```python
# Load the training data and the public test set for the fps feature representation
X_fps_train = pd.read_csv('X_fps_train.csv', index_col = 0) # inputs of the training set
y_train = pd.read_csv('y_train.csv', index_col = 0).to_numpy() # outputs of the training set

X_fps_public_test = pd.read_csv('X_fps_public_test.csv', index_col = 0) # inputs of the public test set
y_public_test = pd.read_csv('y_public_test.csv', index_col = 0) # outputs of the public test set

X_fps_private_test = pd.read_csv('X_fps_private_test.csv', index_col = 0) # inputs of the private test set
```

```python
# X_fps_train is a 975 * 1024 dataframe. 
# Entries (i,j) correspond to the j'th dimension of the observation i
X_fps_train
```

```python
#y_train contains the regression value
y_train[:10]
```

```python
# X_fps_public_test is the array of public test inputs for the fps featurisation, of the same format as X_train. The objective is to predict the output
X_fps_public_test
```

```python
# The training data and the public test set for the embed representation - note the labels are the same
X_embed_train = pd.read_csv('X_embed_train.csv', index_col = 0) # inputs of the training set
X_embed_public_test = pd.read_csv('X_embed_public_test.csv', index_col = 0) # inputs of the public test set
```

```python
# X_embed_train is a 975 * 224 dataframe. 
# Entries (i,j) correspond to the j'th dimension of the observation i
X_embed_train
```

# Classification with knn and prediction on the test set

```python
# Example with a k-nearest neighbours classifier for the fps featurisation
clf = KNeighborsRegressor(n_neighbors = 3)
clf.fit(X_fps_train, y_train)
print(f'MSE of 3-nn on the training set: {np.mean((y_train-clf.predict(X_fps_train))**2):.3f}') # evaluate on the training set
print(f'MSE of 3-nn on the public test set: {np.mean((y_public_test-clf.predict(X_fps_public_test))**2):.3f}') # evaluate on the training set
```

```python
# Compute knn predictions on the test inputs
y_pred = clf.predict(X_fps_private_test)
y_pred.shape
```

# Export in csv format

```python
# Export the predictions on the test data in csv format
prediction = pd.DataFrame(y_pred, columns=['Prediction'])
prediction.index.name='Id'
prediction.to_csv('myprediction.csv') # export to csv file

# The csv file should be of the form
#Id, Prediction
#0, 7.11
#1, 8.22
#...
#72, 9.33
```
